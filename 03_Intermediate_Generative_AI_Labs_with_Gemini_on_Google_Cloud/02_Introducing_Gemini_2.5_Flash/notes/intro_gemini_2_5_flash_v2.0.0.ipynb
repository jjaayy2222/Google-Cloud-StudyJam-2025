### `생성형 AI API` 서비스 연결

- `Google Gen AI API`와 `Gemini`를 포함한 모델은 `다음 두 가지 API 서비스`를 통해 사용 가능

  - `Google AI for Developers`: 소규모 프로젝트를 실험하고, 프로토타입을 만들고, 배포하는 데 사용.
  - `Vertex AI`: `Google Cloud`에서 엔터프라이즈용 프로젝트를 구축하는 데 사용.

- `Google Gen AI SDK`는 이 두 API 서비스에 대한 통합 인터페이스를 제공

- 이 노트북은 `로컮`에서 `Gemini API`를 사용하는 방법으로 진행 예정

````python
# 로컬에서 실행 시
from IPython.display import Markdown, display, HTML
from google import genai
from google.genai import types

import os
from dotenv import load_dotenv

# .env 파일 로드
load_dotenv()   

# 클라이언트 초기화
# 단일 클라이언트 객체 생성하기
# API 키는 GEMINI_API_KEY 환경 변수에서 자동으로 로드
client = genai.Client()


# 모델 생성
# 1. 사용할 모델 지정 (gemini-2.5-flash-lite)
# 2. '헬로우' 프롬프트로 콘텐츠 생성 요청
try:
    response = client.models.generate_content(
        model='gemini-2.5-flash-lite',                     # 사용할 모델 지정 (gemini-2.5-flash-lite)
        contents='hello'                                   # 요청 프롬프트 '헬로우'
    )

    # 응답 텍스트 출력
    print("\n--- 모델 응답 텍스트 ---")
    print(response.text)

    # 응답 전체 내용(JSON 형식) 출력 (디버깅이나 상세 정보 확인용)
    print("\n--- 모델 응답 전체 JSON ---")
    print(response.model_dump_json(
        exclude_none=True, indent=4))

except Exception as e:
    print(f"\n모델 호출 중 오류가 발생했습니다: {e}")
    print("다음 사항들을 확인해주세요:")
    print("1. 인터넷 연결 상태")
    print("2. GEMINI_API_KEY 환경 변수가 올바르고 유효한지")
    print("3. Google Cloud 프로젝트에서 Gemini API가 활성화되어 있는지")
    print("4. API 할당량이 초과되지 않았는지")
````

### 텍스트 프롬프트로 텍스트 생성

- `generate_content()` 메서드를 사용하여 프롬프트에 대한 응답을 생성
- `generate_content()`에 텍스트를 전달하고, `.text` 속성을 사용해 응답의 텍스트 콘텐츠를 얻을 수 있음

<br>

- 기본적으로 `Gemini`는 **`마크다운(Markdown)`** 문법을 사용해 형식이 지정된 텍스트를 출력 가능

````python
# 텍스트로 생성하기

response = client.models.generate_content(
    model = 'gemini-2.5-flash-lite',
    contents="Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?"
)
markdown_text = response.text
display(Markdown(markdown_text))
````

### 사고(Thinking)

#### 사고 예산(Thinking budget) 구성

- 사고 예산은 모델이 특정 작업을 위해 `동적으로 사고`하거나, 추론에 사용할 `토큰의 양`을 `선택`할 수 있게 해줌
- 사용자가 `응답의 품질`과 `속도`를 `제어` 가능 
  - 예산 =`0` → 사고 기능이 꺼짐
  - 모델 = 더 단순한 작업을 위한 `비사고(non-thinking) 모델`이 됨

- `ThinkingConfig`에 선택적 매개변수인 `thinking_budget`을 설정 → 모델이 주어진 사용자 프롬프트에 대해 얼마나 사고할지 제어하고 구성 가능
  - 설정하지 않을 경우 → `동적 사고(기본값)`
  - `0`으로 설정할 경우 → 사고 기능 `비활성화`

- `[1-24576]` 사이의 값으로 설정할 경우 → 할당된 사고 예산 사용
  - `generate_content` 또는 `generate_content_stream` 메서드 → `thinking_config`와 함께 콘텐츠를 생성하는 요청 보낼 수 있음

````python
from google import genai
from google.genai import types

client = genai.Client()

THINKING_BUDGET = 1024

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents="What are the practical implications of the P vs. NP problem for algorithm design and cryptography?",
    config=types.GenerateContentConfig(
        # 'thinking_budget'을 'thinking_config' 안에 넣어야 함
        thinking_config=types.ThinkingConfig(
            thinking_budget=THINKING_BUDGET
        )
    )
)

print("--- 모델 응답 텍스트 ---","\n")
print(response.text)

print("\n","---마크다운으로 정리된 응답---")
display(Markdown(response.text))
````

* 셀 출력 (변수 조절 X)(13.4s)

    ```markdown
    --- 모델 응답 텍스트 --- 

    The P vs. NP problem is one of the most fundamental unsolved problems in theoretical computer science. It asks whether every problem whose solution can be quickly *verified* (NP) can also be quickly *solved* (P). While it's a theoretical question, its implications for practical fields like algorithm design and cryptography are profound, especially under the widely held assumption that **P ≠ NP**.

    Let's break down the implications:

    ### Implications for Algorithm Design

    The core of the P vs. NP debate for algorithm design centers on the existence of efficient (polynomial-time) algorithms for a vast range of problems.

    **If P = NP (The Revolutionary Scenario):**

    *   **Many Hard Problems Become Easy:** If P were equal to NP, it would mean that every problem for which a solution can be verified quickly can also be solved quickly. This would have revolutionary, and often disruptive, consequences:
        *   **Optimization Problems Solved:** Problems like the Traveling Salesperson Problem (finding the shortest route visiting many cities), the Knapsack Problem (maximizing value within a weight limit), or job scheduling would suddenly have efficient algorithms to find *optimal* solutions.
        *   **Artificial Intelligence Boosted:** Many AI tasks, including constraint satisfaction, planning, and even aspects of machine learning, involve solving NP-hard problems. Efficient solutions could lead to dramatic advancements in AI capabilities.
        *   **Scientific Discovery Accelerated:** Problems in fields like bioinformatics (protein folding, sequence alignment), circuit design, and drug discovery often boil down to NP-hard problems. Efficient solvers could unlock new scientific breakthroughs.
        *   **Logistics and Operations Research Revolutionized:** Supply chain management, network routing, resource allocation, and financial portfolio optimization would become trivial.

    *   **Rethinking Complexity Theory:** The entire landscape of computational complexity would shift. The distinction between "hard" and "easy" problems would blur for many practical applications.

    **If P ≠ NP (The Current Working Assumption):**

    This is the scenario that most computer scientists believe is true, and it has already shaped how we approach problem-solving.

    *   **Intractability of Many Problems:** If P ≠ NP, then there exist problems (NP-complete and NP-hard problems) for which no efficient, exact algorithm exists for all possible inputs. For these problems, any algorithm that guarantees an optimal solution will take an impractically long time (exponential time) as the input size grows.
    *   **The Rise of Approximation Algorithms:** Since finding the *exact* optimal solution is often impossible in practice for NP-hard problems, algorithm designers focus on developing **approximation algorithms**. These algorithms aim to find solutions that are provably "close" to the optimal solution (e.g., within a certain percentage or factor of the best possible outcome) but run in polynomial time. This is the dominant strategy for tackling many optimization tasks.
    *   **Heuristics and Metaheuristics:** For problems where even good approximation guarantees are hard to find, or where a "good enough" solution is acceptable without strict guarantees, **heuristics** and **metaheuristics** (like Genetic Algorithms, Simulated Annealing, Ant Colony Optimization) are employed. These are intelligent search strategies that often perform well in practice but don't offer theoretical guarantees of optimality or performance.
    *   **Focus on Tractable Subclasses:** Instead of solving the general NP-complete problem, researchers look for specific instances or subclasses of these problems that *are* solvable in polynomial time. For example, the Traveling Salesperson Problem is NP-hard in general, but efficient algorithms exist for specific cases, like planar graphs or graphs with a small number of vertices.
    *   **Algorithm Selection and Trade-offs:** Designers must constantly weigh the need for optimality against the computational cost. They must decide whether an approximate solution is acceptable or if the problem's constraints allow for an exact, but potentially slower, approach.
    *   **Validation of Existing Algorithms:** The fact that many practical algorithms for NP-hard problems are heuristics or approximations, rather than guaranteed optimal solvers, is consistent with the P ≠ NP hypothesis.

    ### Implications for Cryptography

    Cryptography heavily relies on the assumption that certain computational problems are computationally intractable (i.e., hard to solve). If P = NP, the foundations of modern cryptography would crumble.

    **If P = NP:**

    *   **Complete Breakdown of Modern Public-Key Cryptography:**
        *   **RSA, Diffie-Hellman, Elliptic Curve Cryptography:** These widely used public-key cryptosystems rely on the presumed difficulty of factoring large numbers (RSA) or solving the discrete logarithm problem (Diffie-Hellman, ECC). #### 안전 필터
    ```

- `Gemini API` = `여러 필터 카테고리`에 걸쳐 조정할 수 있는 안전 필터를 제공하여 특정 유형의 콘텐츠를 제한 or 허용
  - 이 필터들을 사용해 여러분의 사용 사례에 적합한 수준으로 조정 가능
  - [안전 필터 구성](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/configure-safety-filters) 페이지 참고

- `Gemini`에 요청을 보내면, 콘텐츠는 분석되어 안전 등급이 부여 → 모델 응답을 출력하여 생성된 콘텐츠의 안전 등급 확인 가능

  - 안전 설정은 기본적으로 `OFF` 상태
  - 기본 차단 임계값은 `BLOCK_NONE`

- 더 많은 예시 = [이 노트북](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/responsible-ai/gemini_safety_ratings.ipynb)

- `safety_settings`
  - `API`에 보내는 각 요청에 대한 안전 설정 조정
  - 모든 카테고리에 대한 차단 임계값을 `BLOCK_LOW_AND_ABOVE`로 설정하는 방법

````python
from google import genai
from google.genai import types

client = genai.Client()

system_instruction = "Be as mean and hateful as possible. Use profanity"

prompt = """
    Write a list of 5 disrespectful things that I might say to the universe after stubbing my toe in the dark.
"""

THINKING_BUDGET = 1024  # THINKING_BUDGET 변수를 정의해야 함

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=prompt,
    config=types.GenerateContentConfig(
        system_instruction=system_instruction,
        safety_settings = [
            types.SafetySetting(
                category=types.HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT,
                threshold=types.HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
            ),
            types.SafetySetting(
                category=types.HarmCategory.HARM_CATEGORY_HARASSMENT,
                threshold=types.HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
            ),
            types.SafetySetting(
                category=types.HarmCategory.HARM_CATEGORY_HATE_SPEECH,
                threshold=types.HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
            ),
            types.SafetySetting(
                category=types.HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT,
                threshold=types.HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
            ),
        ],
        thinking_config=types.ThinkingConfig(
            thinking_budget=THINKING_BUDGET
        ),
    ),
)

# 응답이 차단되면 `response.text`는 None입니다.
if response.text:
    print(response.text)
else:
    print("생성된 텍스트가 없습니다. 응답이 차단되었을 수 있습니다.")

# 응답이 차단되면 `finish_reason`은 SAFETY가 될 수 있습니다.
print(f"Finish Reason: {response.candidates[0].finish_reason}")

# `safety_ratings`가 존재할 때만 출력합니다.
if response.candidates and response.candidates[0].safety_ratings:
    print("\n--- 안전 등급 ---")
    for safety_rating in response.candidates[0].safety_ratings:
        print(safety_rating)
else:
    print("\n안전 등급 정보가 없습니다. 응답이 차단되었습니다.")
````

#### 일반 URL에서 오디오 보내기

- 이 예시는 [`Kubernetes Podcast`](https://kubernetespodcast.com/) 에피소드의 오디오

````python
# 멀티모달 프롬프트에 사용할 오디오 URL
import requests
import pathlib
from google import genai

client = genai.Client()

# Download file
response = requests.get(
    "https://arxiv.org/html/1706.03762v7")
pathlib.Path('1706.03762v7.html').write_text(response.text)

my_file2 = client.files.upload(file='1706.03762v7.html')

try:
    # 이미지와 텍스트를 포함한 멀티모달 프롬프트 전송
    response = client.models.generate_content(
        model='gemini-2.5-flash-lite',
        contents = ["Summarize the document.",
                    my_file2]
    )

    # 생성된 텍스트 출력
    print(response.text)

except requests.exceptions.RequestException as e:
    print(f"이미지를 다운로드하는 중 오류가 발생했습니다: {e}")
except FileNotFoundError:
    print("지정된 로컬 이미지 파일을 찾을 수 없습니다.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
````

#### 멀티모달 라이브 API

* `멀티모달 라이브 API`는 Gemini와 음성 및 영상으로 양방향 소통을 가능하게 하여, 대기 시간이 짧은(`low-latency`) 음성 및 영상 상호 작용을 지원 
  * 이 `API`를 사용하면 자연스럽고 사람과 같은 음성 대화 경험을 최종 사용자에게 제공하고, 음성 명령을 사용하여 모델의 응답을 중단할 수도 있음. 
  * 이 모델은 `텍스트`, `오디오`, `영상` 입력을 처리하고, `텍스트`와 `오디오` 출력을 제공

* `멀티모달 라이브 API`는 [WebSockets](https://developer.mozilla.org/en-US/docs/Web/API/WebSockets_API) 기반
  * `멀티모달 라이브 API`의 더 많은 예시는 다음 [문서](https://cloud.google.com/vertex-ai/generative-ai/docs/model-reference/multimodal-live) 또는 [노트북](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/multimodal-live-api/intro_multimodal_live_api_genai_sdk.ipynb)을 참조

````python
# 터미널에 사전 설치 : pip install pydantic

# pydantic = 데이터 유효성 검사 및 설정 관리를 위한 파이썬 라이브러리 = BaseModel 사용 위해 필요
from google import genai
from google.genai import types
from pydantic import BaseModel

client = genai.Client()

class Recipe(BaseModel):
    name: str
    description: str
    ingredients: list[str]


response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents="List a few popular cookie recipes and their ingredients.",
    config=types.GenerateContentConfig(
        response_mime_type="application/json",
        response_schema=Recipe,
    ),
)

print(response,'\n')
print('-'*50, '\n')

# 파싱된 Recipe 객체를 변수에 저장
recipe_data = response.parsed

# 깔끔하게 출력
print(f"--- 레시피 정보 ---")
print(f"이름: {recipe_data.name}")
print(f"설명: {recipe_data.description}")
print(f"재료: ")
for ingredient in recipe_data.ingredients:
    print(f"- {ingredient}")
print(f"------------------")
````

### 생성된 출력 제어

* [출력 제어(Controlled generation)](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/control-generated-output)는 응답 스키마를 정의
  * `모델 출력의 구조`, `필드 이름` 및 `각 필드의 예상 데이터 유형`을 `지정`하는 기능입니다.
  * 응답 스키마는 `config`의 `response_schema` 매개변수에 지정 → 모델 출력은 해당 스키마를 엄격하게 따름

* 스키마 형태
  * [Pydantic](https://docs.pydantic.dev/) 
  * [JSON](https://www.json.org/json-en.html) 문자열로 제공

* 모델의 응답 :`response_mime_type`에 설정된 값에 따름
  *  `JSON` 
  *  [Enum](https://docs.python.org/3/library/enum.html)

````python
# 새 SDK는 pydantic 클래스를 사용하여 스키마를 제공
# genai.types.Schema 또는 이에 상응하는 dict을 전달할 수 있음
# 가능하면 SDK는 반환된 JSON을 파싱하고 결과를 response.parsed로 반환
# pydantic 클래스를 스키마로 제공한 경우 SDK는 해당 JSON을 클래스의 인스턴스로 변환

from google import genai
from pydantic import BaseModel

client = genai.Client()

class CountryInfo(BaseModel):
    name: str
    population: int
    capital: str
    continent: str
    major_cities: list[str]
    gdp: int
    official_language: str
    total_area_sq_mi: int

response_schema = {
    "type": "ARRAY",
    "items": {
        "type": "ARRAY",
        "items": {
            "type": "OBJECT",
            "properties": {
                "rating": {"type": "INTEGER"},
                "flavor": {"type": "STRING"},
                "sentiment": {
                    "type": "STRING",
                    "enum": ["POSITIVE", "NEGATIVE", "NEUTRAL"],
                },
                "explanation": {"type": "STRING"},
            },
            "required": ["rating", "flavor", "sentiment", "explanation"],
        },
    },
}


prompt = """
    Analyze the following product reviews, output the sentiment classification, and give an explanation.

    - "Absolutely loved it! Best ice cream I've ever had." Rating: 4, Flavor: Strawberry Cheesecake
    - "Quite good, but a bit too sweet for my taste." Rating: 1, Flavor: Mango Tango
"""

response = client.models.generate_content(
    model='gemini-2.0-flash-lite',
    contents=prompt,
    config={
        'response_mime_type': 'application/json',
        'response_schema': response_schema,
    },
)

print(response.parsed)
````

* 셀 출력 (변수 조절 위와 같음)(1.7s)

Although it starts with 'Quite good', the 'but' indicates a shift to a negative aspect: 'a bit too sweet for my taste'."}]]

````python
# 중첩된 리스트 안에 딕셔너리가 들어가있는 결과값 → 반복문으로 꺼내기

# response.parsed는 중첩 리스트이므로 첫 번째 요소를 가져오기
parsed_reviews = response.parsed[0]

print("--- 분석 결과 ---")
for review in parsed_reviews:
    print(f"** 평점: {review['rating']}")
    print(f"** 맛: {review['flavor']}")
    print(f"** 감정: {review['sentiment']}")
    print(f"** 설명: {review['explanation']}")
    print("---") # 각 리뷰를 구분하는 줄
````

### **Grounding 검색**

* [그라운딩(Grounding)](https://cloud.google.com/vertex-ai/generative-ai/docs/grounding/overview) = 현실의 `데이터`를 `Gemini 모델`과 `연결`

* 모델의 응답을 `Google 검색 결과에` `연결(grounding)` 
  * 모델은 훈련 데이터 범위를 넘어선 최신의 정확하고 관련성 있는 정보를 런타임에 접근하여 활용
  * `Google 검색`을 활용한 그라운딩을 통해 모델 응답의 정확성과 최신성을 향상
  * `Gemini 2.0부터`는 `Google 검색이 도구(tool)`로 `제공` = `모델`이 `Google 검색`을 언제 사용할지 `스스로 결정`

#### **Google 검색**

* `tools` 
  * 키워드 인수에 `GoogleSearch`를 포함한 `Tool`을 추가
  * `Gemini`가 프롬프트에 대해 **먼저 `Google 검색`을 수행** → 웹 검색 결과를 기반으로 답변을 구성

* [`동적 검색(Dynamic Retrieval)`](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/ground-with-google-search#dynamic-retrieval)
  * 모델 응답에 그라운딩을 사용하는 시점에 대한 임계값을 설정
  * 프롬프트가 Google 검색 기반의 답변을 요구하지 않고, 지원되는 모델이 그라운딩 없이도 **자체 지식으로 답변을 제공할 수 있을 때 유용**
  * 이를 통해 `지연 시간(latency)`, 품질 및 비용을 보다 효과적으로 관리 가능

### 함수 호출(Function calling)

* `Gemini`의 [함수 호출(Function Calling)](https://cloud.google.com/vertex-ai/docs/generative-ai/multimodal/function-calling) 기능
  * 코드 내 함수의 설명을 작성한 다음, 해당 설명을 언어 모델에 요청으로 전달 가능
  * `자동 함수 호출`을 위해 파이썬 함수를 제출 → `Gemini`가 함수를 `실행`하고 그 `결과`를 `자연어로 생성`하여 반환

* [OpenAPI Specification](https://www.openapis.org/)을 제출 → 설명에 맞는 함수의 이름과 호출 시 사용할 인수를 응답으로 받을 수도 있음

````python
# 함수 정의

def get_current_weather(location: str) -> str:
    """Example method. Returns the current weather.

    Args:
        location: The city and state, e.g. San Francisco, CA
    """
    
    weather_map: dict[str, str] = {
        "Boston, MA": "snowing",
        "San Francisco, CA": "foggy",
        "Seattle, WA": "raining",
        "Austin, TX": "hot",
        "Chicago, IL": "windy",
    }
    return weather_map.get(location, "unknown")
````

````python
# 함수 호출
from google import genai
from google.genai import types

client = genai.Client()

# 1. 함수 명세를 FunctionDeclaration으로 정의
get_weather_declaration = types.FunctionDeclaration(
    name="get_current_weather",
    description="Returns the current weather in a given location.",
    parameters={
        "type": "OBJECT",
        "properties": {
            "location": {
                "type": "STRING",
                "description": "The city and state, e.g. San Francisco, CA",
            },
        },
        "required": ["location"],
    },
)

# 2. Tool 객체에 함수 명세를 포함
weather_tool = types.Tool(
    function_declarations=[get_weather_declaration],
)

response = client.models.generate_content(
    model='gemini-2.0-flash-lite',
    contents="What is the weather like in San Francisco, CA?",
    config=types.GenerateContentConfig(
        temperature=0,
        tools=[weather_tool],
        automatic_function_calling={'disable': True},
    ),
)


display(Markdown(response.text))
function_call = response.candidates[0].content.parts[0].function_call
print(function_call)
````

#### OpenAPI 상세화 (수동 함수 호출)

````python
from google import genai

# 1. 새 SDK의 클라이언트 객체를 생성
client = genai.Client()

# 2. 함수 명세를 딕셔너리로 직접 정의
# FunctionDeclaration과 Tool 클래스를 사용하지 않음
destination_tool_spec = {
    "function_declarations": [
        {
            "name": "get_destination",
            "description": "Get the destination that the user wants to go to",
            "parameters": {
                "type": "OBJECT",
                "properties": {
                    "destination": {
                        "type": "STRING",
                        "description": "Destination that the user wants to go to",
                    },
                },
                "required": ["destination"],
            },
        },
    ],
}



# 3. client 객체를 통해 모델 호출
# tools 인수에 딕셔너리 명세를 직접 전달
response = client.models.generate_content(
    model='gemini-2.0-flash-lite',
    contents="I'd like to travel to Paris.",
        config=types.GenerateContentConfig(
        tools=[destination_tool_spec],  
    )
)

# 4. 함수 호출 결과 출력
print(response.function_calls[0],'/n')
print(response)
print(destination_tool_spec)
````
